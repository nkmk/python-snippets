```python
print(type(l1_l2_and))
```

```python
l1_l2_and_list = list(l1_l2_and)
print(l1_l2_and_list)
```

```python
print(type(l1_l2_and_list))
```

```python
print(len(l1_l2_and))
```

```python
l1_l2_l3_and = set(l1) & set(l2) & set(l3)
print(l1_l2_l3_and)
```
