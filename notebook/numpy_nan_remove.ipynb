```python
import numpy as np
```

```python
print(np.__version__)
```

```python
a = np.genfromtxt('data/src/sample_nan.csv', delimiter=',')
print(a)
```

```python
print(np.isnan(a))
```

```python
print(~np.isnan(a))
```

```python
print(a[~np.isnan(a)])
```

```python
print(np.isnan(a).any(axis=1))
```

```python
print(~np.isnan(a).any(axis=1))
```

```python
print(a[~np.isnan(a).any(axis=1), :])
```

```python
print(a[~np.isnan(a).any(axis=1)])
```

```python
a[1, 0] = np.nan
a[1, 3] = np.nan
print(a)
```

```python
print(np.isnan(a).all(axis=1))
```

```python
print(~np.isnan(a).all(axis=1))
```

```python
print(a[~np.isnan(a).all(axis=1)])
```

```python
a = np.genfromtxt('data/src/sample_nan.csv', delimiter=',')
print(a)
```

```python
print(np.isnan(a))
```

```python
print(np.isnan(a).any(axis=0))
```

```python
print(~np.isnan(a).any(axis=0))
```

```python
print(a[:, ~np.isnan(a).any(axis=0)])
```

```python
a[2, 2] = np.nan
print(a)
```

```python
print(np.isnan(a).all(axis=0))
```

```python
print(~np.isnan(a).all(axis=0))
```

```python
print(a[:, ~np.isnan(a).all(axis=0)])
```
