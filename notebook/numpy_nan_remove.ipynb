```python
import numpy as np
```

```python
arr = np.genfromtxt('data/src/sample_nan.csv', delimiter=',')
print(arr)
```

```python
print(np.isnan(arr))
```

```python
print(arr[~np.isnan(arr)])
```

```python
print(np.isnan(arr).any(axis=1))
```

```python
print(~np.isnan(arr).any(axis=1))
```

```python
print(arr[~np.isnan(arr).any(axis=1), :])
```

```python
print(arr[~np.isnan(arr).any(axis=1)])
```

```python
print(~np.isnan(arr).any(axis=0))
```

```python
print(arr[:, ~np.isnan(arr).any(axis=0)])
```
