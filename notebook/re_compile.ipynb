```python
import re
```

```python
s = 'aaa@xxx.com bbb@yyy.net ccc@zzz.org'
```

```python
print(re.match(r'([a-z]+)@([a-z]+)\.com', s))
```

```python
print(re.sub(r'([a-z]+)@([a-z]+)\.com', 'NEW_ADDRESS', s))
```

```python
p = re.compile(r'([a-z]+)@([a-z]+)\.com')
```

```python
print(p)
```

```python
print(type(p))
```

```python
print(p.match(s))
```

```python
print(p.sub('NEW_ADDRESS', s))
```
