```python
import pandas as pd
```

```python
df = pd.DataFrame({'age': [24, 42], 'state': ['NY', 'CA'], 'point': [64, 92]},
                  index=['Alice', 'Bob'])
```

```python
print(df)
```

```python
for column_name in df:
    print(type(column_name))
    print(column_name)
    print('======\n')
```

```python
for column_name in df.__iter__():
    print(type(column_name))
    print(column_name)
    print('======\n')
```

```python
for column_name, item in df.iteritems():
    print(type(column_name))
    print(column_name)
    print('~~~~~~')
    
    print(type(item))
    print(item)
    print('------')
    
    print(item['Alice'])
    print(item[0])
    print(item.Alice)
    print('======\n')
```

```python
for index, row in df.iterrows():
    print(type(index))
    print(index)
    print('~~~~~~')
    
    print(type(row))
    print(row)
    print('------')
    
    print(row['point'])
    print(row[2])
    print(row.point)
    print('======\n')
```

```python
for row in df.itertuples():
    print(type(row))
    print(row)
    print('------')
    
    print(row[3])
    print(row.point)
    print('======\n')
```

```python
for row in df.itertuples(name=None):
    print(type(row))
    print(row)
    print('------')
    
    print(row[3])
    print('======\n')
```
