```python
import numpy as np
```

```python
a_1d = np.arange(10)
print(a_1d)
```

```python
print(a_1d[3])
```

```python
a_2d = np.arange(12).reshape(3, 4)
print(a_2d)
```

```python
print(a_2d[0, 2])
```

```python
a_3d = np.arange(24).reshape(2, 3, 4)
print(a_3d)
```

```python
print(a_3d[0, 1, 2])
```

```python
print(a_3d[1, [True, False, True], ::2])
```
