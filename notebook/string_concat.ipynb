```python
s1 = 'aaa'
s2 = 'bbb'

i = 100
f = 0.25
```

```python
# s = s1 + i
# TypeError: must be str, not int
```

```python
s = s1 + '_' + str(i) + '_' + s2 + '_' + str(f)
print(s)
```

```python
s = s1 + '_' + format(i, '05') + '_' + s2 + '_' + format(f, '.5f')
print(s)
```

```python
s = '{}_{:05}_{}_{:.5f}'.format(s1, i, s2, f)
print(s)
```

```python
l = ['aaa', 'bbb', 'ccc']
```

```python
s = ','.join(l)
print(s)
```
