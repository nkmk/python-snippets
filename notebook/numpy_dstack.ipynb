```python
import numpy as np
```

```python
a1 = np.ones((3, 4), int)
print(a1)
```

```python
a2 = np.full((3, 4), 2)
print(a2)
```

```python
print(np.dstack([a1, a2]))
```

```python
print(np.dstack([a1, a2]).shape)
```

```python
print(np.concatenate([a1.reshape(3, 4, 1), a2.reshape(3, 4, 1)], 2))
```
