```python
l = ['Alice', 'Bob', 'Charlie']
```

```python
for name in l:
    print(name)
```

```python
for i, name in enumerate(l):
    print(i, name)
```

```python
for i, name in enumerate(l, 1):
    print(i, name)
```

```python
for i, name in enumerate(l, 42):
    print(i, name)
```

```python
for i, name in enumerate(l, 1):
    print('{:03}_{}'.format(i, name))
```
