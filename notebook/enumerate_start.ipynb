```python
l = ['Alice', 'Bob', 'Charlie']
```

```python
for name in l:
    print(name)
```

```python
for i, name in enumerate(l):
    print(i, name)
```

```python
for i, name in enumerate(l, 1):
    print(i, name)
```

```python
for i, name in enumerate(l, 42):
    print(i, name)
```

```python
for i, name in enumerate(l, 1):
    print(f'{i:03}_{name}')
```

```python
step = 3
for i, name in enumerate(l):
    print(i * step, name)
```
