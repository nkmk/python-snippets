```python
import numpy as np
```

```python
arr = np.array([0, 1, 2])
print(arr)
```

```python
arr_2d = np.array([[0, 1, 2], [3, 4, 5]])
print(arr_2d)
```

```python
arr_2d_sqrt = np.sqrt(arr_2d)
print(arr_2d_sqrt)
```

```python
arr_1 = np.array([[1, 2], [3, 4]])
arr_2 = np.array([[1, 2, 3], [4, 5, 6]])

arr_product = np.dot(arr_1, arr_2)
print(arr_product)
```
