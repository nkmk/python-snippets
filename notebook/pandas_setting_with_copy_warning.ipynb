```python
df = pd.DataFrame({'a': [0, 1, 2], 'b': [3, 4, 5]}, index=['x', 'y', 'z'])
print(df)
```

```python
print(df.loc['x':'y']['a'])
```

```python
df.loc['x':'y']['a'] = 100
```

```python
print(df)
```

```python
print(df.loc[['x', 'y']]['a'])
```

```python
df.loc[['x', 'y']]['a'] = 100
print(df)
```

```python
df.loc['x':'y', 'a'] = 100
print(df)
```

```python
df.loc[['x', 'y'], 'a'] = 0
print(df)
```

```python
df = pd.DataFrame({'a': [0, 1, 2], 'b': [3, 4, 5]}, index=['x', 'y', 'z'])
print(df)
```
