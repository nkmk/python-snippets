```python
import pandas as pd
```

```python
s = pd.Series([1, 2, 3], index=['ONE', 'TWO', 'THREE'])
print(s)
```

```python
print(s.rename({'ONE': 'a', 'THREE': 'c'}))
```

```python
print(s.rename(str.lower))
```

```python
print(s.add_prefix('X_'))
```

```python
print(s.add_suffix('_X'))
```

```python
s.index = ['a', 'b', 'c']
print(s)
```
