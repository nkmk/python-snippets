```python
df = pd.DataFrame({'X': [0, 1, 2], 'Y': [3, 4, 5]}, index=['A', 'B', 'C'])
print(df)
```

```python
df_T = df.T
print(df_T)
```

```python
df_transpose = df.transpose()
print(df_transpose)
```

```python
df.at['A', 'X'] = 100
print(df)
```

```python
print(df_T)
```

```python
print(df_transpose)
```

```python
df_mix = pd.DataFrame({'col_int': [0, 1, 2], 'col_float': [0.1, 0.2, 0.3]}, index=['A', 'B', 'C'])
print(df_mix)
```

```python
df_mix_T = df_mix.T
print(df_mix_T)
```

```python
df_mix.at['A', 'col_int'] = 100
print(df_mix)
```

```python
print(df_mix_T)
```

```python
df = pd.DataFrame({'X': [0, 1, 2], 'Y': [3, 4, 5]}, index=['A', 'B', 'C'])
print(df)
```

```python
df_T_copy = df.T.copy()
print(df_T_copy)
```

```python
df_transpose_copy = df.transpose(copy=True)
print(df_transpose_copy)
```

```python
df.at['A', 'X'] = 100
print(df)
```

```python
print(df_T_copy)
```

```python
print(df_transpose_copy)
```
